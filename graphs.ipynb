```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np

sns.set(style="whitegrid")
# configure for ieeetran paper


concurrent_level = [2, 3, 5]
test_name = ["bursty", "medium", "heavy"]


def plot_concurrent_level(conc_level, test_name):
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.bar(test_name, concurrent_level, color="skyblue")

    ax.set_xlabel("Test Name", fontsize=14)
    ax.set_ylabel("Concurrent Level", fontsize=14)
    ax.set_title("Concurrent Level by Test Name", fontsize=16)
    ax.grid(axis="y", linestyle="--", alpha=0.7)

    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()


plot_concurrent_level(concurrent_level, test_name)
```
